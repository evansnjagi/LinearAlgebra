# Choose useful data

```python
# Modules
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Load the data
filepath = "../data/Mall Customers.xlsx"

df = pd.read_excel(filepath)
df.head()
```

## Choose numerical features: Age, Annual income and Spending score.

Each row represents one customer as a vector:

$x_i = (Age, Income, Spending)$

The customers live as a 3D feature space. For example:

$x_1 = (19, 15, 39)$

```python
# Get the required features
features = ["Age", "Annual Income (k$)", "Spending Score (1-100)"]

X = df[features]

X.head()
```

```python
# Convert to matrix form
X_matrix = X.to_numpy()
print("Matrix shape: ", X_matrix.shape)
print(X_matrix[:5])
```

The X(new dataset) is now a matrix with 200 vectors. We can simply say that $X \in \mathbb{R}^{200\times3}$

## Comparing the first two customers

```python
# Selecting customers
customer_1 = X_matrix[0]
customer_2 = X_matrix[1]

print("Customer 1: ", customer_1)
print("Customer 2: ", customer_2)
```

Next step: Compute the distance between the two customers

```python
# Distance between customer_1 and customer_2

distance = np.linalg.norm(customer_2 - customer_1)

print("Distance: ", round(distance, 2))
```

Large distance means different customers while small distance means similar customers.

# Visualize Customers

```python
# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"])
plt.title("Customer Income Vs. Spending score")
plt.xlabel("Annual Income ($)")
plt.ylabel("Spending score [0 - 100]")

# Save the file
save_path = "/workspaces/LinearAlgebra/mall_customer_segmentaiton/images/income_vs_spending_scatter_plot.png"
plt.savefig(save_path)
plt.show()
```

In the scatter plot, there are distinct observable clusters. We can actually segment customers using the annual income feature as well as the annual income.
